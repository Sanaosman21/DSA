```python
# Task 1: Find Maximum Number
arr = [3, 7, 2, 9, 5]
maximum=arr[0]
for num in arr:
    if (num>maximum):
        maximum=num
    else:
        maximum=maximum
print("maximum number is:",maximum)               
```

```python
# Task 2: Count frequency
arr = [1, 1, 2, 3, 2, 1]
freq={}
for num in arr:
    if num in freq:
        freq[num]+=1
    else:
        freq[num]=1
print(freq)
```

```python
# Problem: Highest Frequency Element
arr = [1, 1, 2, 3, 2, 1,2]
freq={}
for num in arr:
    if num in freq:
        freq[num]+=1
    else:
        freq[num]=1
max_freq=0
answer=None
for key ,value in freq.items():
    if value >max_freq:
        max_freq=value
        answer=key
print(answer)
```
