```python
#square of sorted array
#first square the then apply the logic 

def sortedSquares(nums):
        siz=len(nums)
        pos=[]
        neg=[]
        #separate negative and positive 
        for num in nums:
            if num<0:
                neg.append(num)
            else:
                pos.append(num)

        #case 1 no negatives 
        if len(neg)==0:
            return[x*x for x in pos]

        #case 2 no positives 
        if len(pos)==0:
            res=[x*x for x in neg]
            res.reverse(neg)
            return res
        #case 3 both exist 
        neg=[x*x for x in neg][::-1]
        pos=[x*x for x in pos]
        n,m=len(neg) ,len(pos)
        res=[]
        i=j=0
        while(i<m and j<n):
             if(neg[i]<=pos[j]):
                  res.append(neg[i])
                  i=i+1
             else:
                  res.append(pos[j])
                  j=j+1
        while(i<n):
             res.append(neg[i])
             i+=1
        while(j<m):
             res.append(pos[j])
             j+=1
        return res
nums=[-4,-1,0,3,9]
print(sortedSquares(nums))

                   
```
